```python
import soccerdata as sd
import pandas as pd
```

```python
from soccerdata import FBref
```

```python
# Initialize the Scraper for Top 5 Leagues
seasons = [f"{year}-{year+1}" for year in range(1995, 2025)]
fbref = sd.FBref(leagues="Big 5 European Leagues Combined", seasons=seasons)
```

```python
#Download the data
stats=fbref.read_player_season_stats(stat_type='standard')
```

```python
stats.head()
```

```python
temp_df=stats.reset_index()
columns_to_keep = [
    ("league", ""),
    ("season", ""),
    ("team", ""),
    ("player", ""),
    ("Playing Time", "MP")
]
```

```python
final_df = temp_df[columns_to_keep]
final_df.columns = ["league", "season", "team", "player", "MP"]
```

```python
final_df.head()
```

```python
final_df['MP']= pd.to_numeric(final_df['MP'], errors='coerce'). fillna(0)
```

```python
final_df = final_df.sort_values(by =['player','season'])
```

```python
final_df['Total_Career_Apps']= final_df.groupby('player')['MP'].cumsum()
```

```python
final_df['Previous_Appearances']= final_df['Total_Career_Apps']-final_df['MP']
```

```python
final_df
```

```python
target_season="2425"
target_team ="Manchester City"

roster = final_df[(final_df['team'] == target_team) & (final_df['season'] == target_season) & (final_df['MP'] >= 5)]
```

```python
roster
```

```python
avg_experience = roster['Previous_Appearances'].mean()
```
