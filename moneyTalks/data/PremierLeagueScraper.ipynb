```python
# import the required library
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
from io import StringIO
import time
```

```python
seasons =["2013-2014", "2014-2015","2015-2016","2016-2017","2017-2018","2018-2019","2019-2020","2020-2021","2022-2023","2023-2024","2024-2025"]
base_url = "https://fbref.com/en/comps/9/{0}/{0}-Premier-League-Stats"
```

```python
# instantiate a Chrome options object
options = webdriver.ChromeOptions()

# set the options to use Chrome in headless mode
options.add_argument("--headless=new")
options.add_argument("--window-size=1920,1080")
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
options.add_argument(f'user-agent={user_agent}')

# initialize an instance of the chrome driver (browser)
driver = webdriver.Chrome(options=options)

with pd.ExcelWriter("EPL_Master_Stats.xlsx") as writer:# visit your target site
    for season in seasons:
        target_url = base_url.format(season)
        driver.get(target_url)

#driver.get('https://fbref.com/en/comps/9/2024-2025/2024-2025-Premier-League-Stats')
        html_content =driver.page_source
        tables=pd.read_html(StringIO(html_content), match="Squad")
#First table on the list
        df=tables[0]

        df.to_excel(writer, sheet_name=season,index=False)
        time.sleep(5)

# release the resources allocated by Selenium and shut down the browser
driver.quit()
```
